## Graphing with Data 

```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
%config InlineBackend.figure_format = 'retina'
```

```python
seaborn.set_style(style='whitegrid')
```

```python
# Import the data from the URL as a pandas Dataframe assign it to the variable df
# object assigned to the variable df
df = pandas.read_csv('https://data.medicaid.gov/sites/default/files/uploaded_resources/2023-child-and-adult-health-care-quality-measures.csv')
```

```python
# Display the first 5 rows of the dataframe
df.head()
```

```python
# Columns in the data file
df.columns
```

```python
seaborn.scatterplot(
    data=df,
    x='Measure Abbreviation',
    y='Median',
);
```

```python
seaborn.scatterplot(
    data=df,
    x='Measure Abbreviation',
    y='Median',
    hue='Measure Type',
    style='Measure Type',


).set(
    title='Measure and Success Care',
    xlabel='Measure [name]',
    ylabel='Median [% of care]'
);
```

```python
seaborn.scatterplot(
    data=df,
    x='Median',
    y='Domain',
    hue='State',
    style='Measure Type',
).set(
    title='Measure and Success Care',
    xlabel='Median [% of care]',
    ylabel='Domain'
);
```

```python
seaborn.displot(
    df,
    x='Median'
).set(
    title='Median Distribution of Care',
    xlabel='Median',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Median',
    bins=range(0, 100, 10),
    hue='Domain',
).set(
    title='Median Distribution of Care',
    xlabel='Median',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Median',
    bins=range(0, 100, 10),
    col='State',
    hue='Domain',
    multiple='stack',
).set(
    title='Median Distribution of Care',
    xlabel='Median',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Median',
    y='Domain',
    cbar=True
).set(
    title='Domain and Median',
    xlabel='Median',
    ylabel='Domain'
);
```

```python
seaborn.catplot(
    data=df,
    x='Measure Type',
    y='Median',
    kind='box'
).set(
    title='Measure Type and Median',
    xlabel='Measure Type',
    ylabel='Median'
);
```

```python
seaborn.catplot(
    data=df,
    x='Measure Type',
    y='Median',
    hue='Domain',
    kind='box'
).set(
    title='Measure Type and Median by Domain',
    xlabel='Measure Type',
    ylabel='Median'
);
```

```python
seaborn.catplot(
    data=df,
    x='Measure Type',
    y='Median',
    hue='State',
    kind='boxen'
).set(
    title='Measure Type and Median by State',
    xlabel='Measure Type',
    ylabel='Median'
);
```

```python
seaborn.catplot(
    data=df,
    x='Measure Type',
    y='Median',
    hue='Domain',
    kind='violin'
).set(
    title='Measure Type and Median by Domain',
    xlabel='Measure Type',
    ylabel='Median'
);
```

```python
seaborn.catplot(
    data=df,
    x='Domain',
    kind='count'
).set(
    title='Domain Frequency',
    xlabel='Domain',
    ylabel='Frequency'
);
```

```python
seaborn.catplot(
    data=df,
    x='Domain',
    hue='Measure Type',
    kind='count'
);
```

```python
seaborn.catplot(
    data=df,
    x='Measure Type',
    y='Median',
    hue='Domain',
    kind='point'
).set(
    title='Domain and Median by Measure Type',
    xlabel='Domain',
    ylabel='Median'
);
```

## Quiz Questions:

### Seaborn Visualization Guide

#### Basic Import and Setup
* Import Seaborn: `import seaborn as sns`
* Set style: `sns.set_style("whitegrid")`

#### Scatter Plot
* Basic: `sns.scatterplot(data=df, x='x', y='y')`
* With regression: `sns.regplot(data=df, x='x', y='y')`

#### Line Plot
* Basic: `sns.lineplot(data=df, x='x', y='y')`
* Multiple lines: `sns.lineplot(data=df, x='x', y='y', hue='category')`

#### Histogram
* Single: `sns.histplot(data=df, x='x')`
* Stacked: `sns.histplot(data=df, x='x', hue='y', multiple="stack")`

#### Heatmap
* Basic: `sns.heatmap(df.corr())`
* With annotations: `sns.heatmap(df.corr(), annot=True)`

#### Box Plot
* Basic: `sns.boxplot(data=df, x='x', y='y')`
* With categories: `sns.boxplot(data=df, x='x', y='y', hue='category')`

#### Violin Plot
* Basic: `sns.violinplot(data=df, x='x', y='y')`
* Split by category: `sns.violinplot(data=df, x='x', y='y', hue='category', split=True)`

#### Bar Plot
* Basic: `sns.barplot(data=df, x='x', y='y')`
* With error bars: `sns.barplot(data=df, x='x', y='y', ci=95)`

#### Plot Customization
* Title: `ax.set_title('My Title')`
* Save plot: `plt.savefig('output.png')`

#### Axis Labels
* X-axis: `ax.set_xlabel('X Label')`
* Y-axis: `ax.set_ylabel('Y Label')`

#### Color and Scale
* Color palette: `sns.set_palette("Blues")`
* Log scale: `plt.yscale('log')`

#### Regression Plot
* Basic: `sns.regplot(data=df, x='x', y='y')`
* With confidence interval: `sns.regplot(data=df, x='x', y='y', ci=95)`
